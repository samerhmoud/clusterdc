# Comparing KDE local vs global bandwidth

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# Generate synthetic data
np.random.seed(42)
data = np.vstack([
    np.random.multivariate_normal([2, 2], [[0.1, 0], [0, 0.1]], size=10000),
    np.random.multivariate_normal([5, 5], [[0.5, 0.2], [0.2, 0.5]], size=20000)
])

# Determine meshgrid extent with 20% margin
x_min, x_max = data[:, 0].min(), data[:, 0].max()
y_min, y_max = data[:, 1].min(), data[:, 1].max()
x_margin = 0.2 * (x_max - x_min)
y_margin = 0.2 * (y_max - y_min)

x_min -= x_margin
x_max += x_margin
y_min -= y_margin
y_max += y_margin

# Create grid for KDE evaluation
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
query_points = np.vstack([xx.ravel(), yy.ravel()]).T

# Define kernel functions
kernel_funcs = {
    "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
    "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
    "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
    "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
    "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
}

# Global Bandwidth KDE
def global_kde(data, query_points, bandwidth, kernel_func):
    densities = np.zeros(len(query_points))
    for i, q in enumerate(query_points):
        diffs = data - q  # Differences between data and query point
        norms = np.linalg.norm(diffs, axis=1)  # Euclidean distances
        scaled_diffs = norms / bandwidth  # Scale by global bandwidth
        kernel_values = kernel_func(scaled_diffs) / bandwidth
        densities[i] = np.sum(kernel_values) / len(data)
    return densities

# Optimized Local Bandwidth KDE
def optimized_local_kde(data, query_points, k, kernel_func):
    # Fit NearestNeighbors once
    nbrs = NearestNeighbors(n_neighbors=k).fit(data)
    distances, _ = nbrs.kneighbors(query_points)  # Precompute distances for all query points
    local_bandwidths = distances[:, -1]  # Bandwidth for each query point

    # Compute densities using the precomputed distances and bandwidths
    densities = np.zeros(len(query_points))
    for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
        if h > 0:  # Avoid division by zero
            scaled_diffs = distances[i] / h  # Scale by local bandwidth
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(data)
    return densities

# User selects kernel type
kernel_type = "gaussian"  # Change this value to select a different kernel (e.g., "epanechnikov", "biweight", etc.)
kernel_func = kernel_funcs.get(kernel_type, None)
if kernel_func is None:
    raise ValueError(f"Unsupported kernel type: {kernel_type}")

# Compute KDE for both methods
global_bandwidth = 0.5
global_densities = global_kde(data, query_points, global_bandwidth, kernel_func)
local_densities = optimized_local_kde(data, query_points, k=5, kernel_func=kernel_func)

# Reshape densities for plotting
global_densities = global_densities.reshape(xx.shape)
local_densities = local_densities.reshape(xx.shape)

# Plot Global Bandwidth KDE
plt.figure(figsize=(20, 6))

plt.subplot(1, 3, 1)
plt.contourf(xx, yy, global_densities, levels=50, cmap='viridis')
#plt.scatter(data[:, 0], data[:, 1], s=5, color='white', marker='+')
plt.colorbar(label='Density')
plt.title(f'Global Bandwidth KDE (h = 0.5, kernel = {kernel_type})')
plt.xlabel('X')
plt.ylabel('Y')

# Plot Local Bandwidth KDE
plt.subplot(1, 3, 2)
plt.contourf(xx, yy, local_densities, levels=50, cmap='viridis')
#plt.scatter(data[:, 0], data[:, 1], s=5, color='white', marker='+')
plt.colorbar(label='Density')
plt.title(f'Local Bandwidth KDE (k = 10, kernel = {kernel_type})')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 3)
plt.scatter(data[:, 0], data[:, 1], s=5, color='k')
plt.title('Simulated Points')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import ParameterGrid
from sklearn.neighbors import NearestNeighbors
from tqdm import tqdm
from sklearn.datasets import make_moons, make_circles
from joblib import Parallel, delayed

class ClusterDC:
    def __init__(self, data, user_percentages=[5, 10, 15, 20], kernel_types=None, alpha_values=None):
        self.data = data
        self.user_percentages = user_percentages
        self.kernel_types = kernel_types or ["gaussian", "epanechnikov", "biweight", "triangular", "cosine", "silverman", "scott"]
        self.alpha_values = alpha_values or [0.5, 1.0, 2.0]
        self.grid = None
        self.results_df = None
        self.best_params = None

    def create_grid(self, resolution=100):
        x_min, x_max = self.data[:, 0].min(), self.data[:, 0].max()
        y_min, y_max = self.data[:, 1].min(), self.data[:, 1].max()
        x_margin = 0.1 * (x_max - x_min)
        y_margin = 0.1 * (y_max - y_min)

        x_min -= x_margin
        x_max += x_margin
        y_min -= y_margin
        y_max += y_margin

        xx, yy = np.meshgrid(np.linspace(x_min, x_max, resolution), np.linspace(y_min, y_max, resolution))
        self.grid = np.vstack([xx.ravel(), yy.ravel()]).T

    @staticmethod
    def silverman_bandwidth(data):
        std = np.std(data, axis=0)
        iqr = np.subtract(*np.percentile(data, [75, 25], axis=0))
        return 0.9 * np.minimum(std, iqr / 1.34) * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def scott_bandwidth(data):
        std = np.std(data, axis=0)
        return std * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def global_kde(data, query_points, bandwidth, kernel_func):
        diffs = data[:, None, :] - query_points[None, :, :]
        norms = np.linalg.norm(diffs, axis=2)
        scaled_diffs = norms / bandwidth
        kernel_values = kernel_func(scaled_diffs) / bandwidth
        densities = np.sum(kernel_values, axis=0) / data.shape[0]
        return densities

    def evaluate_kde(self, k, kernel, alpha=1.0, return_densities=False):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }

        if kernel == "silverman":
            global_bandwidth = np.mean(self.silverman_bandwidth(self.data))
            densities = self.global_kde(self.data, self.grid, global_bandwidth, kernel_funcs["gaussian"])
            return densities if return_densities else np.mean(np.log(densities + 1e-9))

        if kernel == "scott":
            global_bandwidth = np.mean(self.scott_bandwidth(self.data))
            densities = self.global_kde(self.data, self.grid, global_bandwidth, kernel_funcs["gaussian"])
            return densities if return_densities else np.mean(np.log(densities + 1e-9))

        nbrs = NearestNeighbors(n_neighbors=k).fit(self.data)
        distances, _ = nbrs.kneighbors(self.grid)
        local_bandwidths = distances[:, -1] * alpha

        def compute_density(q, h):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_funcs[kernel](scaled_diffs) / h
            return np.sum(kernel_values) / len(self.data)

        densities = Parallel(n_jobs=-1)(
            delayed(compute_density)(q, h) for q, h in zip(self.grid, local_bandwidths)
        )

        densities = np.array(densities)
        return densities if return_densities else np.mean(np.log(densities + 1e-9))

    def fit(self):
        k_values = [int(len(self.data) * (p / 100)) for p in self.user_percentages]
        param_grid = {
            "k": k_values,
            "kernel": self.kernel_types,
            "alpha": self.alpha_values,
        }

        results = []
        for params in tqdm(ParameterGrid(param_grid), desc="Grid Search Progress", unit="params"):
            k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
            log_likelihood = self.evaluate_kde(k, kernel, alpha)
            results.append((k, kernel, alpha, log_likelihood))

        self.results_df = pd.DataFrame(results, columns=["k", "kernel", "alpha", "log_likelihood"])
        self.best_params = max(results, key=lambda x: x[3])

    def plot(self):
        if self.best_params is None:
            raise ValueError("No best parameters found. Run the fit() method first.")

        best_k, best_kernel, best_alpha = self.best_params[:3]
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }

        if best_kernel in ["silverman", "scott"]:
            bandwidth = np.mean(self.silverman_bandwidth(self.data)) if best_kernel == "silverman" else np.mean(self.scott_bandwidth(self.data))
            densities = self.global_kde(self.data, self.grid, bandwidth, kernel_funcs["gaussian"])
        else:
            densities = self.evaluate_kde(best_k, best_kernel, best_alpha, return_densities=True)

        densities = densities.reshape(int(np.sqrt(len(self.grid))), -1)

        plt.figure(figsize=(12, 6))

        # KDE Density Plot
        plt.subplot(1, 2, 1)
        plt.contourf(np.linspace(self.grid[:, 0].min(), self.grid[:, 0].max(), densities.shape[1]),
                     np.linspace(self.grid[:, 1].min(), self.grid[:, 1].max(), densities.shape[0]), densities, levels=50, cmap='viridis')
        plt.colorbar(label='Density')
        plt.title(f"2D KDE with {best_kernel.capitalize()} Bandwidth (k={best_k}, alpha={best_alpha})")
        plt.xlabel("X")
        plt.ylabel("Y")

        # Scatter Plot of Data Points
        plt.subplot(1, 2, 2)
        plt.scatter(self.data[:, 0], self.data[:, 1], s=5, color='black')
        plt.title("Scatter Plot of Data Points")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.tight_layout()
        plt.show()

# Example usage
np.random.seed(42)
data = np.vstack([
    np.random.multivariate_normal([2, 2], [[0.1, 0], [0, 0.1]], size=100),
    np.random.multivariate_normal([5, 5], [[0.5, 0.2], [0.2, 0.5]], size=200)
])


clusterdc = ClusterDC(data, alpha_values=[0.1, 0.5, 2.0])
clusterdc.create_grid()
clusterdc.fit()
clusterdc.plot()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
from scipy.spatial import KDTree
import matplotlib.pyplot as plt
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import networkx as nx
import warnings
import copy

warnings.filterwarnings('ignore')
np.random.seed(42)

class ClusterDC:
    def __init__(self, x_points, y_points, precomputed_density=None, precomputed_grid=None, 
                 levels=50, min_point=1, border_fraction=0.5, selection_method='first_gap'):
        self.x_points = x_points
        self.y_points = y_points
        self.levels = levels
        self.min_point = min_point
        self.border_fraction = border_fraction
        self.selection_method = selection_method
        self.precomputed_density = precomputed_density
        self.precomputed_grid = precomputed_grid  # Accept precomputed grid
        self.xy = np.array([[x_points[i], y_points[i]] for i in range(len(x_points))])

    def create_grid(self):
        deltaX = (np.max(self.x_points) - np.min(self.x_points)) * self.border_fraction
        deltaY = (np.max(self.y_points) - np.min(self.y_points)) * self.border_fraction
        xmin = np.min(self.x_points) - deltaX
        xmax = np.max(self.x_points) + deltaX
        ymin = np.min(self.y_points) - deltaY
        ymax = np.max(self.y_points) + deltaY
        xx, yy = np.mgrid[xmin:xmax:self.num_points * 1j, ymin:ymax:self.num_points * 1j]
        return xx, yy

    def evaluate_kde(self):
        if self.precomputed_density is not None and self.precomputed_grid is not None:
            # Use precomputed density and grid
            xx, yy = np.unique(self.precomputed_grid[:, 0]), np.unique(self.precomputed_grid[:, 1])
            xx, yy = np.meshgrid(xx, yy)
            return xx, yy, self.precomputed_density.reshape(xx.shape)
        else:
            raise ValueError("Precomputed density or grid is missing, and no KDE computation method available.")

    def create_contours(self, xx, yy, f):
        fig, ax = plt.subplots(figsize=(10, 10))
        cset = ax.contour(xx, yy, f, levels=self.levels)
        plt.close()
        polygons, density_values, density_contours = [], [], []
        for j, level in enumerate(cset.levels):
            for seg in cset.allsegs[j]:
                polygons.append([seg[:, 0], seg[:, 1]])
                density_values.append(level)
                density_contours.append(j)
        return polygons, density_values, density_contours

    def make_node_dic(self, polygons, density_values, density_contours):
        dic_node = {ind: {'level': density_contours[ind], 'density': density_values[ind], 'childrens': [], 'parent': [], 'token': []} for ind in range(len(polygons))}
        for ind in range(len(polygons)):
            if polygons[ind][0][0] == polygons[ind][0][-1] and polygons[ind][1][0] == polygons[ind][1][-1]:
                for ind_sub in range(len(polygons)):
                    if polygons[ind_sub][0][0] == polygons[ind_sub][0][-1] and polygons[ind_sub][1][0] == polygons[ind_sub][1][-1]:
                        if dic_node[ind_sub]['level'] == dic_node[ind]['level'] + 1 and dic_node[ind_sub]['density'] > dic_node[ind]['density']:
                            poly = Polygon([(polygons[ind][0][index], polygons[ind][1][index]) for index in range(len(polygons[ind][0]))])
                            poly_sub = Polygon([(polygons[ind_sub][0][index], polygons[ind_sub][1][index]) for index in range(len(polygons[ind_sub][0]))])
                            if poly.contains(poly_sub):
                                dic_node[ind]['childrens'].append(ind_sub)
                                dic_node[ind_sub]['parent'].append(ind)

        # Assign 'root' as a parent for top-level nodes without a parent
        for node in dic_node:
            if not dic_node[node]['parent']:
                dic_node[node]['parent'] = ['*']

        dic_node['*'] = {'level': 0, 'density': 0, 'childrens': [node for node in dic_node if '*' in dic_node[node]['parent']], 'parent': [], 'token': []}
        return dic_node



    def find_leaves(self, dic_node):
        return [ind for ind in dic_node.keys() if ind != '*' and not dic_node[ind]['childrens']]

    def find_plateau(self, dic_node):
        ind_plateau = []
        for ind in dic_node.keys():
            if len(dic_node[ind]['childrens']) > 1:
                ind_plateau.append(ind)
        return ind_plateau

    def check_node_unvalid(self, xy, polygons, ind_node, min_point):
        if not isinstance(ind_node, int):  # Skip non-numeric nodes like 'root'
            return False
        poly_current = Polygon([(polygons[ind_node][0][index], polygons[ind_node][1][index]) for index in range(len(polygons[ind_node][0]))])
        nb_point_inside = sum(1 for point in xy if Point(point[0], point[1]).within(poly_current))
        return nb_point_inside < min_point

    def trim_unvalid_leaves(self, xy, dic_node, polygons, min_point):
        for level in sorted(set(node['level'] for node in dic_node.values() if isinstance(node, dict)), reverse=True):
            nodes_at_level = [node for node in dic_node if isinstance(node, int) and dic_node[node]['level'] == level]
            for node in nodes_at_level:
                parent = dic_node[node]['parent'][0]
                if parent == 'root':  # Skip trimming for top-level nodes
                    continue
                if self.check_node_unvalid(xy, polygons, node, min_point):
                    dic_node[parent]['childrens'] = [child for child in dic_node[parent]['childrens'] if child != node]
                    del dic_node[node]
        return dic_node


    def trim_chain_nodes(self, dic_node):
        """
        Removes unnecessary chain nodes by connecting parent nodes directly to children.
        """
        for level in sorted(set(node['level'] for node in dic_node.values() if isinstance(node, dict)), reverse=True):
            nodes_at_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for node in nodes_at_level:
                if not dic_node[node]['parent']:
                    continue  # Skip nodes with no parent
                parent = dic_node[node]['parent'][0]
                if parent == '**':  # Skip trimming for the root node
                    continue
                if len(dic_node[node]['childrens']) == 1:
                    child = dic_node[node]['childrens'][0]
                    if parent in dic_node:  # Ensure parent exists before modifying
                        dic_node[parent]['childrens'] = [child if n == node else n for n in dic_node[parent]['childrens']]
                    dic_node[child]['parent'] = [parent]
                    del dic_node[node]
        return dic_node



    def find_closest_point(self, point, points):
        """
        Finds the closest point to the input point from a set of input points.
        Parameters:
            point (array): An array representing the point whose closest point is to be found.
            points (array): An array of points.
        Returns:
            (tuple): A tuple with the closest point, the index of closest point, and the distance to closest point.
        """
        # Create a KDTree using the input points
        tree = KDTree(points)
        # Find the distance and index of the closest point to the input point using the KDTree
        dist, idx = tree.query(point)
        # Return the closest point, the index of the closest point, and the distance to the closest point
        return points[idx], idx, dist
    
    def clean_dic_after_choice(self, ind_selection, dic_node):
        """
        Cleans the input dictionary by removing nodes and updating tokens after the selection of points.
        Parameters:
            ind_selection (list): A list of selected point indices.
            dic_node (dict): A dictionary of nodes, each representing a contour.
        Returns:
            dic_node (dict): the cleaned dictionary.
        """
        # Ensure the root node exists
        if '*' not in dic_node:
            dic_node['*'] = {'level': 0, 'density': 0, 'childrens': [], 'parent': [], 'token': []}

        # Find the leaves in the dictionary
        list_leaves = self.find_leaves(dic_node)
        # Add the leaf indices to the token list of the leaf nodes
        for leaf in list_leaves:
            dic_node[leaf]['token'].append(leaf)

        # Iterate over the levels of the tree in reverse order
        max_level = max(node['level'] for node in dic_node.values() if isinstance(node, dict))
        for level in range(max_level, -1, -1):
            nodes_at_level = [node for node in dic_node if isinstance(node, int) and dic_node[node]['level'] == level]
            for node in nodes_at_level:
                if not dic_node[node]['parent']:  # Skip nodes with no parent
                    continue
                token_to_keep = [token for token in dic_node[node]['token'] if token in ind_selection]
                ind_parent = dic_node[node]['parent'][0]
                if ind_parent != '*':
                    if len(dic_node[node]['childrens']) > 1:
                        dic_node[node]['token'] = token_to_keep
                        if not token_to_keep:
                            del dic_node[node]
                            dic_node[ind_parent]['childrens'] = [ind for ind in dic_node[ind_parent]['childrens'] if ind != node]
                    else:
                        if not token_to_keep:
                            del dic_node[node]
                            dic_node[ind_parent]['childrens'] = [ind for ind in dic_node[ind_parent]['childrens'] if ind != node]
                        else:
                            dic_node[ind_parent]['token'] += dic_node[node]['token']
        # Remove unnecessary chain nodes
        dic_node = self.trim_chain_nodes(dic_node)
        # Update the token list of the root node
        dic_node['*']['token'] = ind_selection
        return dic_node
   
    def debug_missing_parents(dic_node):
        for node in dic_node:
            if not dic_node[node]['parent']:
                print(f"Node {node} is missing a parent.")


    def compute_density_peak(self, xy, polygons, ind_peak, density):
        peak_center, peak_center_density, valid_indices = [], [], []
        for ind in ind_peak:
            poly_current = Polygon([(polygons[ind][0][index], polygons[ind][1][index]) for index in range(len(polygons[ind][0]))])
            inside_points = [density[i] for i, point in enumerate(xy) if Point(point[0], point[1]).within(poly_current)]
            if inside_points:
                max_density_index = np.argmax(inside_points)
                peak_center_density.append(inside_points[max_density_index])
                peak_center.append(xy[max_density_index])
                valid_indices.append(ind)  # Keep track of valid indices
        return peak_center, peak_center_density, valid_indices

    def create_graph(self, polygons, density_values, dic_node):
        G = nx.Graph()
        for node in dic_node.keys():
            if not dic_node[node]['parent']:
                continue  # Skip nodes without a parent
            parent = dic_node[node]['parent'][0]
            G.add_node(node)
            if isinstance(parent, str):  # Handle 'root' or other string parents
                G.add_node(parent)
                G.add_edge(node, parent, weight=(0, np.random.uniform(0, 1)))  # Assign default weight for non-numeric parent
            else:
                G.add_edge(node, parent, weight=(density_values[parent], np.random.uniform(0, 1)))
        return G


    def path_max(self, G, s, t, current_depth):
        if len(list(G.edges.data())) == 1:
            return list(G.edges.data())[0][2]['weight'][0]
        if s == t:
            return 0
        weight = np.array([w[0] for u, v, w in G.edges(data='weight')])
        rand_weight = np.array([w[1] for u, v, w in G.edges(data='weight')])
        median = np.median(weight)
        rand_median = np.median(rand_weight)
        G_k = nx.Graph()
        G_k.add_nodes_from(G)
        if len(np.unique(weight)) == 1:
            return median
        for edge in list(G.edges.data()):
            if edge[2]['weight'][0] > median:
                G_k.add_edge(edge[0], edge[1], weight=edge[2]['weight'])
            if edge[2]['weight'][0] == median:
                if edge[2]['weight'][1] >= np.random.uniform(0, 1):
                    G_k.add_edge(edge[0], edge[1], weight=edge[2]['weight'])
        if t in list(nx.node_connected_component(G_k, s)):
            return self.path_max(G_k.subgraph(nx.node_connected_component(G_k, s)), s, t, current_depth + 1)
        else:
            S = [G_k.subgraph(c) for c in nx.connected_components(G_k)]
            G_bar_k = nx.Graph()
            x, y = None, None
            for u, G_u in enumerate(S):
                G_bar_k.add_node(u)
                if s in list(G_u.nodes()):
                    x = u
                if t in list(G_u.nodes()):
                    y = u
                for v, G_v in enumerate(S):
                    if u != v:
                        maximum_weight = None
                        maximum_link = [None, None]
                        for node_u in list(G_u.nodes()):
                            for node_v in list(G_v.nodes()):
                                if node_v in list(G.neighbors(node_u)):
                                    w = G[node_u][node_v]['weight'][0]
                                    rand_value = G[node_u][node_v]['weight'][1]
                                    if maximum_weight is None:
                                        maximum_weight = w
                                        maximum_link[0] = node_u
                                        maximum_link[1] = node_v
                                    else:
                                        if w > maximum_weight:
                                            maximum_weight = w
                                            maximum_link[0] = node_u
                                            maximum_link[1] = node_v
                        if maximum_weight is not None:
                            G_bar_k.add_edge(u, v, weight=(maximum_weight, np.random.uniform(0, 1)))
            return self.path_max(G_bar_k, x, y, current_depth + 1)

    def calculate_maximum_flow(self, G, peak_center_density, ind_peak):
        dens_df = pd
        dens_df = pd.DataFrame(peak_center_density, columns=['density'])
        dens_df.index = ind_peak
        dens_df = dens_df.sort_values(by=['density'], ascending=False)
        dens_list = dens_df.to_numpy()
        ind_cand = dens_df.index
        max_flow = []
        for ind, i in enumerate(ind_cand):
            maximum = -np.inf
            for j in ind_cand[:ind]:
                current_depth = 0
                flow = self.path_max(G, i, j, current_depth)
                if flow > maximum:
                    maximum = flow
            if maximum == -np.inf:
                if ind == 0:
                    max_flow.append(0)
                else:
                    max_flow.append(np.inf)
            else:
                max_flow.append(maximum)
        return max_flow, dens_list, dens_df

    def get_well_separated_points(self, max_flow, dens_list, dens_df, selection_method):
        nb_values = len(max_flow)
        separability = np.array([1 - max_flow[i] / dens_list[i] for i in range(nb_values)])
        dens_df['separability'] = separability
        dens_df = dens_df.sort_values(by=['separability'], ascending=False)

        if selection_method == 'first_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = - well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the first gap
            first_gap_index = (-well_separated['gap']).argsort()[::-1][1]
            # Return the indices of the well-separated points up to the first gap
            return well_separated.iloc[:first_gap_index + 1]['index'].to_numpy()
        if selection_method == 'second_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the second gap
            second_gap_index = (-well_separated['gap']).argsort()[::-1][2]
            # Return the indices of the well-separated points up to the second gap
            return well_separated.iloc[:second_gap_index + 1, :]['index'].to_numpy()
        if selection_method == 'third_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]       
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the third gap
            third_gap_index = (-well_separated['gap']).argsort()[::-1][3]
            # Return the indices of the well-separated points up to the second gap
            return well_separated.iloc[:third_gap_index + 1, :]['index'].to_numpy()
        if isinstance(selection_method, int):
            # If the selection method is an integer, return the first n indices where n is the selection_method
            return dens_df[dens_df['separability'] > 0].iloc[:selection_method].index.to_numpy()
        if selection_method == 'all':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # If the selection method is 'all', return all the indices where separability is greater than 0
            return dens_df[dens_df['separability'] > 0].index.to_numpy()
        else:
            # If the inputed selection method is invalid, raise an error message
            raise ValueError("Invalid selection method. Must be 'first_gap', 'second_gap', 'third_gap', an integer or 'all'.")

    def get_point_polygon(self, density, xy, polygon, dic_node):
        """
        Given a set of points and their corresponding density and polygon, the function returns a dictionary containing the points included inside each contour.
        Parameters:
            density (list): List of density values of the points.
            xy (list): A list of the x,y coordinates of data points.
            polygon (list): A list of polygons representing the contours. Each polygon is a list of x and y coordinates of the contour.
            dic_node (dict): A dictionary of nodes, each representing a contour.
        Returns:
            dic_point_node (dict): A dictionary containing the points and their corresponding density and coordinates for each contour.
        """
        dic_point_node = {ind: [] for ind in dic_node.keys()}
        # Iterate through each node in the density tree
        for ind_node in dic_node.keys():
            if ind_node != '*':
                poly_current = Polygon([(polygon[ind_node][0][index], polygon[ind_node][1][index]) for index in range(len(polygon[ind_node][0]))])
                # Iterate through each point, check if it is within the current polygon
                for k, point in enumerate(xy):
                    point_current = Point(point[0], point[1])   
                    if point_current.within(poly_current):
                        dic_point_node[ind_node].append([k, point, density[k]])
        # Add all points to the root of the density tree
        dic_point_node['*'] = [[k, xy[k], density[k]] for k in range(len(xy))]
        # Return the dictionary containing the points and their corresponding density and coordinates for each contour
        return dic_point_node
    
    def assign_points(self, xy, dic_point_node, dic_node):
        assignment = np.array([None for _ in range(len(xy))])
        max_level = max(dic_node[node]['level'] for node in dic_node)
        for level in range(max_level, -1, -1):
            nodes_at_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for node in nodes_at_level:
                if not dic_node[node]['childrens']:
                    for k, _, _ in dic_point_node[node]:
                        assignment[k] = int(dic_node[node]['token'][0])
                else:
                    already_assigned, not_assigned = [], []
                    already_assigned_value, already_assigned_point = [], []
                    not_assigned_point, not_assigned_den = [], []
                    for k, point, density in dic_point_node[node]:
                        if assignment[k] is not None:
                            already_assigned.append(k)
                            already_assigned_value.append(int(assignment[k]))
                            already_assigned_point.append(point)
                        else:
                            not_assigned.append(k)
                            not_assigned_point.append(point)
                            not_assigned_den.append(density)
                    if not_assigned:
                        sorted_unassigned = sorted(zip(not_assigned_den, not_assigned, not_assigned_point), reverse=True)
                        for _, k, point in sorted_unassigned:
                            closest_point = self.find_closest_point(point, already_assigned_point)
                            assignment[k] = already_assigned_value[closest_point[1]]
        return assignment

    def run_clustering(self):
        xx, yy, f = self.evaluate_kde()
        polygons, density_values, density_contours = self.create_contours(xx, yy, f)
        dic_node = self.make_node_dic(polygons, density_values, density_contours)
        dic_node = self.trim_unvalid_leaves(self.xy, dic_node, polygons, self.min_point)
        dic_node = self.trim_chain_nodes(dic_node)
        ind_peak = self.find_leaves(dic_node)
        density = self.precomputed_density if self.precomputed_density is not None else [
            self.get_kde_point(point) for point in self.xy
        ]
        peak_center, peak_center_density, valid_indices = self.compute_density_peak(self.xy, polygons, ind_peak, density)
        G = self.create_graph(polygons, density_values, dic_node)
        max_flow, dens_list, dens_df = self.calculate_maximum_flow(G, peak_center_density, valid_indices)  # Use aligned indices
        ind_selection = list(self.get_well_separated_points(max_flow, dens_list, dens_df, self.selection_method))

        if self.selection_method != 'all':
            indices = [ind for ind, k in enumerate(valid_indices) if k in ind_selection]
            dic_node = self.clean_dic_after_choice(ind_selection, dic_node)
            dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node)
            assignment = self.assign_points(self.xy, dic_point_node, dic_node)
            dic_reassignment = {selection: index for index, selection in enumerate(ind_selection)}
            assignment = [dic_reassignment[selection] for selection in assignment]
            density_info = [density, f, xx, yy]
            print('============================')
            print('Number of clusters: ', len(np.unique(assignment)))
            print('============================')
            return assignment, density_info
        else:
            dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node)
            list_assignment = []
            for j in range(len(ind_selection)):
                dic_node_sub = copy.deepcopy(dic_node)
                ind_selection_sub = ind_selection[:j + 1]
                indices_sub = [ind for ind, k in enumerate(valid_indices) if k in ind_selection_sub]
                dic_node_sub = self.clean_dic_after_choice(ind_selection_sub, dic_node_sub)
                dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node_sub)
                assignment_sub = self.assign_points(self.xy, dic_point_node, dic_node_sub)
                dic_reassignment = {selection: index for index, selection in enumerate(ind_selection_sub)}
                assignment_sub = [dic_reassignment[selection] for selection in assignment_sub]
                list_assignment.append(assignment_sub)
            density_info = [density, f, xx, yy]
            print('============================')
            print('Max. number of clusters: ', len(np.unique(assignment_sub)))
            print('============================')
            return list_assignment, density_info
```

# Make Clusterdc Class

```python
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
from scipy.spatial import KDTree
import matplotlib.pyplot as plt
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import networkx as nx
import warnings
import copy

warnings.filterwarnings('ignore')
np.random.seed(42)

class ClusterDC:
    def __init__(self, x_points, y_points, levels=50, num_points=100, min_point=1, border_fraction=0.5, 
                 bw_method='scott', selection_method='first_gap', bw=None):
        self.x_points = x_points
        self.y_points = y_points
        self.levels = levels
        self.num_points = num_points
        self.min_point = min_point
        self.border_fraction = border_fraction
        self.bw_method = bw_method
        self.selection_method = selection_method
        self.bw = bw
        self.kernel = None
        self.xy = np.array([[x_points[i], y_points[i]] for i in range(len(x_points))])
        if bw_method != 'local':
            self.kernel = gaussian_kde(np.vstack([x_points, y_points]), bw_method)

    def create_grid(self):
        deltaX = (np.max(self.x_points) - np.min(self.x_points)) * self.border_fraction
        deltaY = (np.max(self.y_points) - np.min(self.y_points)) * self.border_fraction
        xmin = np.min(self.x_points) - deltaX
        xmax = np.max(self.x_points) + deltaX
        ymin = np.min(self.y_points) - deltaY
        ymax = np.max(self.y_points) + deltaY
        xx, yy = np.mgrid[xmin:xmax:self.num_points * 1j, ymin:ymax:self.num_points * 1j]
        return xx, yy

    def get_kde_point(self, point):
        if self.bw_method == 'local':
            density = 0
            for i, bwi in enumerate(self.bw):
                density += (1 / (bwi ** 2)) * np.exp(-((point[0] - self.x_points[i]) ** 2 + (point[1] - self.y_points[i]) ** 2) / (2 * (bwi ** 2)))
            return density / len(self.bw)
        else:
            return self.kernel.evaluate(np.array([point[0], point[1]]))

    def evaluate_kde(self, xx, yy):
        density_map = [self.get_kde_point([xi, yi]) for xi, yi in zip(xx.ravel(), yy.ravel())]
        return np.reshape(density_map, xx.shape)

    def create_contours(self, xx, yy, f):
        fig, ax = plt.subplots(figsize=(10, 10))
        cset = ax.contour(xx, yy, f, levels=self.levels)
        plt.close()
        polygons, density_values, density_contours = [], [], []
        for j, level in enumerate(cset.levels):
            for seg in cset.allsegs[j]:
                polygons.append([seg[:, 0], seg[:, 1]])
                density_values.append(level)
                density_contours.append(j)
        return polygons, density_values, density_contours

    def make_node_dic(self, polygons, density_values, density_contours):
        dic_node = {ind: {'level': density_contours[ind], 'density': density_values[ind], 'childrens': [], 'parent': [], 'token': []} for ind in range(len(polygons))}
        for ind in range(len(polygons)):
            if polygons[ind][0][0] == polygons[ind][0][-1] and polygons[ind][1][0] == polygons[ind][1][-1]:
                for ind_sub in range(len(polygons)):
                    if polygons[ind_sub][0][0] == polygons[ind_sub][0][-1] and polygons[ind_sub][1][0] == polygons[ind_sub][1][-1]:
                        if dic_node[ind_sub]['level'] == dic_node[ind]['level'] + 1 and dic_node[ind_sub]['density'] > dic_node[ind]['density']:
                            poly = Polygon([(polygons[ind][0][index], polygons[ind][1][index]) for index in range(len(polygons[ind][0]))])
                            poly_sub = Polygon([(polygons[ind_sub][0][index], polygons[ind_sub][1][index]) for index in range(len(polygons[ind_sub][0]))])
                            if poly.contains(poly_sub):
                                dic_node[ind]['childrens'].append(ind_sub)
                                dic_node[ind_sub]['parent'].append(ind)
        max_level = max(node['level'] for node in dic_node.values())
        for level in range(max_level, 0, -1):
            list_node_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for ind_node in list_node_level:
                poly = Polygon([(polygons[ind_node][0][index], polygons[ind_node][1][index]) for index in range(len(polygons[ind_node][0]))])
                for ind_node_sub in list_node_level:
                    poly_sub = Polygon([(polygons[ind_node_sub][0][index], polygons[ind_node_sub][1][index]) for index in range(len(polygons[ind_node_sub][0]))])
                    if poly.contains(poly_sub) and ind_node != ind_node_sub:
                        if ind_node_sub in dic_node:
                            del dic_node[ind_node_sub]
        children = [ind for ind in dic_node if dic_node[ind]['level'] == 1]
        for child in children:
            dic_node[child]['parent'] = '*'
        dic_node['*'] = {'level': 0, 'density': 0, 'childrens': children, 'parent': ['**'], 'token': []}
        return dic_node

    def find_leaves(self, dic_node):
        return [ind for ind in dic_node.keys() if ind != '*' and not dic_node[ind]['childrens']]

    def find_plateau(self, dic_node):
        ind_plateau = []
        for ind in dic_node.keys():
            if len(dic_node[ind]['childrens']) > 1:
                ind_plateau.append(ind)
        return ind_plateau

    def check_node_unvalid(self, xy, polygons, ind_node, min_point):
        if ind_node == '*':
            return False
        poly_current = Polygon([(polygons[ind_node][0][index], polygons[ind_node][1][index]) for index in range(len(polygons[ind_node][0]))])
        nb_point_inside = sum(1 for point in xy if Point(point[0], point[1]).within(poly_current))
        return nb_point_inside < min_point

    def trim_unvalid_leaves(self, xy, dic_node, polygons, min_point):
        for level in sorted(set(node['level'] for node in dic_node.values()), reverse=True):
            nodes_at_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for node in nodes_at_level:
                parent = dic_node[node]['parent'][0]
                if self.check_node_unvalid(xy, polygons, node, min_point):
                    dic_node[parent]['childrens'] = [child for child in dic_node[parent]['childrens'] if child != node]
                    del dic_node[node]
        return dic_node

    def trim_chain_nodes(self, dic_node):
        for level in sorted(set(node['level'] for node in dic_node.values()), reverse=True):
            nodes_at_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for node in nodes_at_level:
                parent = dic_node[node]['parent'][0]
                if parent == '**':  # Skip trimming for the root node
                    continue
                if len(dic_node[node]['childrens']) == 1:
                    child = dic_node[node]['childrens'][0]
                    if parent in dic_node:  # Ensure parent exists before modifying
                        dic_node[parent]['childrens'] = [child if n == node else n for n in dic_node[parent]['childrens']]
                    dic_node[child]['parent'] = [parent]
                    del dic_node[node]
        return dic_node

    def find_closest_point(self, point, points):
        """
        Finds the closest point to the input point from a set of input points.
        Parameters:
            point (array): An array representing the point whose closest point is to be found.
            points (array): An array of points.
        Returns:
            (tuple): A tuple with the closest point, the index of closest point, and the distance to closest point.
        """
        # Create a KDTree using the input points
        tree = KDTree(points)
        # Find the distance and index of the closest point to the input point using the KDTree
        dist, idx = tree.query(point)
        # Return the closest point, the index of the closest point, and the distance to the closest point
        return points[idx], idx, dist
    
    def clean_dic_after_choice(self, ind_selection, dic_node):
        """
        Cleans the input dictionary by removing nodes and updating tokens after the selection of points.
        Parameters:
            ind_selection (list): A list of selected point indices.
            dic_node (dict): A dictionary of nodes, each representing a contour.
        Returns:
            dic_node (dict): the cleaned dictionary.
        """
        # Find the leaves in the dictionary
        list_leaves = self.find_leaves(dic_node)
        # Add the leaf indices to the token list of the leaf nodes
        for leaf in list_leaves:
            dic_node[leaf]['token'].append(leaf)
        # Find the maximum level of the tree
        max_level = -1
        for node in dic_node.keys():
            if dic_node[node]['level'] > max_level:
                max_level = dic_node[node]['level']
        # Iterate over the levels of the tree in reverse order
        for level in range(max_level, -1, -1):
            list_node_level = []
            # Get the list of nodes at the current level
            for node in dic_node.keys():
                if dic_node[node]['level'] == level:
                    list_node_level.append(node)
            # Iterate over the nodes at the current level
            for node in list_node_level:
                token_to_keep = []
                # Get the tokens that need to be kept
                for token in dic_node[node]['token']:
                    if token in ind_selection:
                        token_to_keep.append(token)
                # Get the parent node of the current node
                ind_parent = dic_node[node]['parent'][0]
                if ind_parent != '**':
                    if len(dic_node[node]['childrens']) > 1:
                        dic_node[node]['token'] = token_to_keep
                        # If the current node has more than one child
                        if len(token_to_keep) == 0:
                            # If there are no tokens to keep, delete the current node
                            if node in dic_node.keys():
                                del dic_node[node]
                            dic_node[ind_parent]['childrens'] = [ind for ind in dic_node[ind_parent]['childrens'] if ind != node]
                        else:
                            # If there are tokens to keep, update the token list of the parent node
                            dic_node[ind_parent]['token'] = [*dic_node[ind_parent]['token'], *token_to_keep] 
                    else:
                        # If the current node has only one child
                        if len(token_to_keep) == 0:
                            # If there are no tokens to keep, delete the current node
                            if node in dic_node.keys():
                                del dic_node[node]
                            dic_node[ind_parent]['childrens'] = [ind for ind in dic_node[ind_parent]['childrens'] if ind != node]              
                        else:
                            # If there are tokens to keep, update the token list of the parent node
                            dic_node[ind_parent]['token'] = [*dic_node[ind_parent]['token'], *dic_node[node]['token']]
        # Remove unnecessary chain nodes
        dic_node = self.trim_chain_nodes(dic_node)
        # Update the token list of the root node
        dic_node['*']['token'] = ind_selection
        # Return the cleaned dictionary
        return dic_node

    def compute_density_peak(self, xy, polygons, ind_peak, density):
        peak_center, peak_center_density = [], []
        for ind in ind_peak:
            poly_current = Polygon([(polygons[ind][0][index], polygons[ind][1][index]) for index in range(len(polygons[ind][0]))])
            inside_points = [density[i] for i, point in enumerate(xy) if Point(point[0], point[1]).within(poly_current)]
            if inside_points:
                max_density_index = np.argmax(inside_points)
                peak_center_density.append(inside_points[max_density_index])
                peak_center.append(xy[max_density_index])
        return peak_center, peak_center_density

    def create_graph(self, polygons, density_values, dic_node):
        G = nx.Graph()
        for node in dic_node.keys():
            parent = dic_node[node]['parent'][0]
            G.add_node(node)
            G.add_node(parent)
            if parent in ['**', '*']:
                G.add_edge(node, parent, weight=(0, np.random.uniform(0, 1)))
            else:
                G.add_edge(node, parent, weight=(density_values[parent], np.random.uniform(0, 1)))
        return G

    def path_max(self, G, s, t, current_depth):
        if len(list(G.edges.data())) == 1:
            return list(G.edges.data())[0][2]['weight'][0]
        if s == t:
            return 0
        weight = np.array([w[0] for u, v, w in G.edges(data='weight')])
        rand_weight = np.array([w[1] for u, v, w in G.edges(data='weight')])
        median = np.median(weight)
        rand_median = np.median(rand_weight)
        G_k = nx.Graph()
        G_k.add_nodes_from(G)
        if len(np.unique(weight)) == 1:
            return median
        for edge in list(G.edges.data()):
            if edge[2]['weight'][0] > median:
                G_k.add_edge(edge[0], edge[1], weight=edge[2]['weight'])
            if edge[2]['weight'][0] == median:
                if edge[2]['weight'][1] >= np.random.uniform(0, 1):
                    G_k.add_edge(edge[0], edge[1], weight=edge[2]['weight'])
        if t in list(nx.node_connected_component(G_k, s)):
            return self.path_max(G_k.subgraph(nx.node_connected_component(G_k, s)), s, t, current_depth + 1)
        else:
            S = [G_k.subgraph(c) for c in nx.connected_components(G_k)]
            G_bar_k = nx.Graph()
            x, y = None, None
            for u, G_u in enumerate(S):
                G_bar_k.add_node(u)
                if s in list(G_u.nodes()):
                    x = u
                if t in list(G_u.nodes()):
                    y = u
                for v, G_v in enumerate(S):
                    if u != v:
                        maximum_weight = None
                        maximum_link = [None, None]
                        for node_u in list(G_u.nodes()):
                            for node_v in list(G_v.nodes()):
                                if node_v in list(G.neighbors(node_u)):
                                    w = G[node_u][node_v]['weight'][0]
                                    rand_value = G[node_u][node_v]['weight'][1]
                                    if maximum_weight is None:
                                        maximum_weight = w
                                        maximum_link[0] = node_u
                                        maximum_link[1] = node_v
                                    else:
                                        if w > maximum_weight:
                                            maximum_weight = w
                                            maximum_link[0] = node_u
                                            maximum_link[1] = node_v
                        if maximum_weight is not None:
                            G_bar_k.add_edge(u, v, weight=(maximum_weight, np.random.uniform(0, 1)))
            return self.path_max(G_bar_k, x, y, current_depth + 1)

    def calculate_maximum_flow(self, G, peak_center_density, ind_peak):
        dens_df = pd
        dens_df = pd.DataFrame(peak_center_density, columns=['density'])
        dens_df.index = ind_peak
        dens_df = dens_df.sort_values(by=['density'], ascending=False)
        dens_list = dens_df.to_numpy()
        ind_cand = dens_df.index
        max_flow = []
        for ind, i in enumerate(ind_cand):
            maximum = -np.inf
            for j in ind_cand[:ind]:
                current_depth = 0
                flow = self.path_max(G, i, j, current_depth)
                if flow > maximum:
                    maximum = flow
            if maximum == -np.inf:
                if ind == 0:
                    max_flow.append(0)
                else:
                    max_flow.append(np.inf)
            else:
                max_flow.append(maximum)
        return max_flow, dens_list, dens_df

    def get_well_separated_points(self, max_flow, dens_list, dens_df, selection_method):
        nb_values = len(max_flow)
        separability = np.array([1 - max_flow[i] / dens_list[i] for i in range(nb_values)])
        dens_df['separability'] = separability
        dens_df = dens_df.sort_values(by=['separability'], ascending=False)

        if selection_method == 'first_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = - well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the first gap
            first_gap_index = (-well_separated['gap']).argsort()[::-1][1]
            # Return the indices of the well-separated points up to the first gap
            return well_separated.iloc[:first_gap_index + 1]['index'].to_numpy()
        if selection_method == 'second_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the second gap
            second_gap_index = (-well_separated['gap']).argsort()[::-1][2]
            # Return the indices of the well-separated points up to the second gap
            return well_separated.iloc[:second_gap_index + 1, :]['index'].to_numpy()
        if selection_method == 'third_gap':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]       
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # Get the index of the third gap
            third_gap_index = (-well_separated['gap']).argsort()[::-1][3]
            # Return the indices of the well-separated points up to the second gap
            return well_separated.iloc[:third_gap_index + 1, :]['index'].to_numpy()
        if isinstance(selection_method, int):
            # If the selection method is an integer, return the first n indices where n is the selection_method
            return dens_df[dens_df['separability'] > 0].iloc[:selection_method].index.to_numpy()
        if selection_method == 'all':
            # Select the well-separated points
            well_separated = dens_df[dens_df['separability'] > 0]
            # Append a row with negative infinity values
            well_separated.loc['*',:] = [0, 0]
            # Calculate the gaps between separability values
            well_separated['gap'] = -well_separated.separability.diff()
            # Reset the index of the DataFrame
            well_separated = well_separated.reset_index(drop = False)
            # Print the cluster center candidates separabilities
            print(well_separated.loc[:,['separability', 'gap']].round(2))
            # If the selection method is 'all', return all the indices where separability is greater than 0
            return dens_df[dens_df['separability'] > 0].index.to_numpy()
        else:
            # If the inputed selection method is invalid, raise an error message
            raise ValueError("Invalid selection method. Must be 'first_gap', 'second_gap', 'third_gap', an integer or 'all'.")

    def get_point_polygon(self, density, xy, polygon, dic_node):
        """
        Given a set of points and their corresponding density and polygon, the function returns a dictionary containing the points included inside each contour.
        Parameters:
            density (list): List of density values of the points.
            xy (list): A list of the x,y coordinates of data points.
            polygon (list): A list of polygons representing the contours. Each polygon is a list of x and y coordinates of the contour.
            dic_node (dict): A dictionary of nodes, each representing a contour.
        Returns:
            dic_point_node (dict): A dictionary containing the points and their corresponding density and coordinates for each contour.
        """
        dic_point_node = {ind: [] for ind in dic_node.keys()}
        # Iterate through each node in the density tree
        for ind_node in dic_node.keys():
            if ind_node != '*':
                poly_current = Polygon([(polygon[ind_node][0][index], polygon[ind_node][1][index]) for index in range(len(polygon[ind_node][0]))])
                # Iterate through each point, check if it is within the current polygon
                for k, point in enumerate(xy):
                    point_current = Point(point[0], point[1])   
                    if point_current.within(poly_current):
                        dic_point_node[ind_node].append([k, point, density[k]])
        # Add all points to the root of the density tree
        dic_point_node['*'] = [[k, xy[k], density[k]] for k in range(len(xy))]
        # Return the dictionary containing the points and their corresponding density and coordinates for each contour
        return dic_point_node
    
    def assign_points(self, xy, dic_point_node, dic_node):
        assignment = np.array([None for _ in range(len(xy))])
        max_level = max(dic_node[node]['level'] for node in dic_node)
        for level in range(max_level, -1, -1):
            nodes_at_level = [node for node in dic_node if dic_node[node]['level'] == level]
            for node in nodes_at_level:
                if not dic_node[node]['childrens']:
                    for k, _, _ in dic_point_node[node]:
                        assignment[k] = int(dic_node[node]['token'][0])
                else:
                    already_assigned, not_assigned = [], []
                    already_assigned_value, already_assigned_point = [], []
                    not_assigned_point, not_assigned_den = [], []
                    for k, point, density in dic_point_node[node]:
                        if assignment[k] is not None:
                            already_assigned.append(k)
                            already_assigned_value.append(int(assignment[k]))
                            already_assigned_point.append(point)
                        else:
                            not_assigned.append(k)
                            not_assigned_point.append(point)
                            not_assigned_den.append(density)
                    if not_assigned:
                        sorted_unassigned = sorted(zip(not_assigned_den, not_assigned, not_assigned_point), reverse=True)
                        for _, k, point in sorted_unassigned:
                            closest_point = self.find_closest_point(point, already_assigned_point)
                            assignment[k] = already_assigned_value[closest_point[1]]
        return assignment

    def run_clustering(self):
        xx, yy = self.create_grid()
        f = self.evaluate_kde(xx, yy)
        polygons, density_values, density_contours = self.create_contours(xx, yy, f)
        dic_node = self.make_node_dic(polygons, density_values, density_contours)
        dic_node = self.trim_unvalid_leaves(self.xy, dic_node, polygons, self.min_point)
        dic_node = self.trim_chain_nodes(dic_node)
        ind_peak = self.find_leaves(dic_node)
        density = [self.get_kde_point(point) for point in self.xy]
        peak_center, peak_center_density = self.compute_density_peak(self.xy, polygons, ind_peak, density)
        G = self.create_graph(polygons, density_values, dic_node)
        max_flow, dens_list, dens_df = self.calculate_maximum_flow(G, peak_center_density, ind_peak)
        ind_selection = list(self.get_well_separated_points(max_flow, dens_list, dens_df, self.selection_method))

        if self.selection_method != 'all':
            indices = [ind for ind, k in enumerate(ind_peak) if k in ind_selection]
            dic_node = self.clean_dic_after_choice(ind_selection, dic_node)
            dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node)
            assignment = self.assign_points(self.xy, dic_point_node, dic_node)
            dic_reassignment = {selection: index for index, selection in enumerate(ind_selection)}
            assignment = [dic_reassignment[selection] for selection in assignment]
            density_info = [density, f, xx, yy]
            print('============================')
            print('Number of clusters: ', len(np.unique(assignment)))
            print('============================')
            return assignment, density_info
        else:
            dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node)
            list_assignment = []
            for j in range(len(ind_selection)):
                dic_node_sub = copy.deepcopy(dic_node)
                ind_selection_sub = ind_selection[:j + 1]
                indices_sub = [ind for ind, k in enumerate(ind_peak) if k in ind_selection_sub]
                dic_node_sub = self.clean_dic_after_choice(ind_selection_sub, dic_node_sub)
                dic_point_node = self.get_point_polygon(density, self.xy, polygons, dic_node_sub)
                assignment_sub = self.assign_points(self.xy, dic_point_node, dic_node_sub)
                dic_reassignment = {selection: index for index, selection in enumerate(ind_selection_sub)}
                assignment_sub = [dic_reassignment[selection] for selection in assignment_sub]
                list_assignment.append(assignment_sub)
            density_info = [density, f, xx, yy]
            print('============================')
            print('Max. number of clusters: ', len(np.unique(assignment_sub)))
            print('============================')
            return list_assignment, density_info

```

```python
def read_file(file_path):
    """
    Reads a file (.xlsx, .xls, or .csv) into a pandas DataFrame.
    
    Parameters:
        file_path (str): Path to the input file.
        
    Returns:
        DataFrame: The data read from the file.
    """
    if file_path.endswith(".xlsx") or file_path.endswith(".xls"):
        return pd.read_excel(file_path)
    elif file_path.endswith(".csv"):
        return pd.read_csv(file_path)
    else:
        raise ValueError("Unsupported file format. Please provide a .xlsx, .xls, or .csv file.")
```

```python
class KDE:
    def __init__(self, data, user_percentages=None, kernel_types=None, alpha_values=None):
        self.data = data
        self.user_percentages = user_percentages if user_percentages is not None else [10]
        self.kernel_types = kernel_types if kernel_types is not None else ["gaussian"]
        self.alpha_values = alpha_values if alpha_values is not None else [2.0]
        self.xx, self.yy = None, None
        self.results_df = None
        self.best_params = None

    def _calculate_bandwidth(self, method):
        if method == "silverman":
            std = np.std(self.data, axis=0)
            iqr = np.subtract(*np.percentile(self.data, [75, 25], axis=0))
            return 0.9 * np.minimum(std, iqr / 1.34) * len(self.data) ** (-1 / (self.data.shape[1] + 4))
        elif method == "scott":
            std = np.std(self.data, axis=0)
            return std * len(self.data) ** (-1 / (self.data.shape[1] + 4))

    def _global_kde(self, query_points, bandwidth, kernel_func):
        densities = np.zeros(len(query_points))
        for i, q in enumerate(query_points):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / bandwidth
            kernel_values = kernel_func(scaled_diffs) / bandwidth
            densities[i] = np.sum(kernel_values) / len(self.data)
        return densities

    def _evaluate_kde(self, query_points, k, kernel, alpha=1.0):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        }

        kernel_func = kernel_funcs.get(kernel, None)
        if not kernel_func:
            raise ValueError(f"Unsupported kernel: {kernel}")

        nbrs = NearestNeighbors(n_neighbors=k).fit(self.data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * alpha

        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(self.data)

        return np.mean(np.log(densities + 1e-9))

    def fit(self):
        x_min, x_max = self.data[:, 0].min(), self.data[:, 0].max()
        y_min, y_max = self.data[:, 1].min(), self.data[:, 1].max()
        x_margin = 0.1 * (x_max - x_min)
        y_margin = 0.1 * (y_max - y_min)

        x_min -= x_margin
        x_max += x_margin
        y_min -= y_margin
        y_max += y_margin

        self.xx, self.yy = np.meshgrid(np.linspace(x_min, x_max, 50), np.linspace(y_min, y_max, 50))
        query_points = np.vstack([self.xx.ravel(), self.yy.ravel()]).T

        k_values = [int(len(self.data) * (p / 100)) for p in self.user_percentages]
        results = []

        for params in tqdm(ParameterGrid({"k": k_values, "kernel": self.kernel_types, "alpha": self.alpha_values}), desc="Grid Search Progress"):
            k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
            log_likelihood = self._evaluate_kde(query_points, k, kernel, alpha)
            results.append((k, kernel, alpha, log_likelihood))

        self.results_df = pd.DataFrame(results, columns=["k", "kernel", "alpha", "log_likelihood"]).sort_values(by="log_likelihood", ascending=False)
        self.best_params = self.results_df.iloc[0]
        return self.results_df

    def plot(self):
        if self.best_params is None:
            raise RuntimeError("You need to call the `fit` method before plotting.")

        query_points = np.vstack([self.xx.ravel(), self.yy.ravel()]).T
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        }

        best_k = int(self.best_params["k"])
        best_kernel = self.best_params["kernel"]
        best_alpha = self.best_params["alpha"]

        kernel_func = kernel_funcs[best_kernel]
        nbrs = NearestNeighbors(n_neighbors=best_k).fit(self.data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * best_alpha

        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(self.data)

        densities = densities.reshape(self.xx.shape)

        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.contourf(self.xx, self.yy, densities, levels=50, cmap="viridis")
        plt.colorbar(label="Density")
        plt.title(f"2D KDE with Gaussian Bandwidth")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.subplot(1, 2, 2)
        plt.scatter(self.data[:, 0], self.data[:, 1], s=5, color="black")
        plt.title("Scatter Plot of Data Points")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.tight_layout()
        plt.show()


```

```python
from sklearn.model_selection import ParameterGrid
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import pandas as pd
from tqdm import tqdm
from sklearn.datasets import make_moons, make_circles
pd.set_option('display.max_rows', None)

data = read_file('training_data.csv').values


# Determine meshgrid extent with 10% margin
x_min, x_max = data[:, 0].min(), data[:, 0].max()
y_min, y_max = data[:, 1].min(), data[:, 1].max()
x_margin = 0.1 * (x_max - x_min)
y_margin = 0.1 * (y_max - y_min)

x_min -= x_margin
x_max += x_margin
y_min -= y_margin
y_max += y_margin

# Create grid for KDE evaluation
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
query_points = np.vstack([xx.ravel(), yy.ravel()]).T

# User specifies percentage values for k
user_percentages = [10]  # Example: User inputs percentages
k_values = [int(len(data) * (p / 100)) for p in user_percentages]

# # Define a parameter grid for hyperparameter tuning
# param_grid = {
#     "k": k_values,  # Number of neighbors as percentages of total samples
#     "kernel": ["gaussian", "epanechnikov", "biweight", "triangular", "cosine", "silverman", "scott"],
#     "alpha": [0.5, 1.0, 2.0],
# }
# Define a parameter grid for hyperparameter tuning
param_grid = {
    "k": k_values,  # Number of neighbors as percentages of total samples
    "kernel": ["gaussian"],
    "alpha": [2.0],
}
# Function to compute Silverman and Scott bandwidths
def silverman_bandwidth(data):
    std = np.std(data, axis=0)
    iqr = np.subtract(*np.percentile(data, [75, 25], axis=0))
    return 0.9 * np.minimum(std, iqr / 1.34) * len(data) ** (-1 / (data.shape[1] + 4))

def scott_bandwidth(data):
    std = np.std(data, axis=0)
    return std * len(data) ** (-1 / (data.shape[1] + 4))

# Global Bandwidth KDE Function
def global_kde(data, query_points, bandwidth, kernel_func):
    densities = np.zeros(len(query_points))
    for i, q in enumerate(query_points):
        diffs = data - q
        norms = np.linalg.norm(diffs, axis=1)
        scaled_diffs = norms / bandwidth
        kernel_values = kernel_func(scaled_diffs) / bandwidth
        densities[i] = np.sum(kernel_values) / len(data)
    return densities

# Modify evaluate_kde to include global bandwidth options
def evaluate_kde(data, query_points, k, kernel, alpha=1.0):
    kernel_funcs = {
        "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
        "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
        "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
        "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
    }
    if kernel == "silverman":
        global_bandwidth = np.mean(silverman_bandwidth(data))
        return np.mean(np.log(global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
    if kernel == "scott":
        global_bandwidth = np.mean(scott_bandwidth(data))
        return np.mean(np.log(global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
    kernel_func = kernel_funcs.get(kernel, None)
    if not kernel_func:
        raise ValueError(f"Unsupported kernel: {kernel}")

    nbrs = NearestNeighbors(n_neighbors=k).fit(data)
    densities = np.zeros(len(query_points))
    distances, _ = nbrs.kneighbors(query_points)
    local_bandwidths = distances[:, -1] * alpha
    for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
        diffs = data - q
        norms = np.linalg.norm(diffs, axis=1)
        scaled_diffs = norms / h
        kernel_values = kernel_func(scaled_diffs) / h
        densities[i] = np.sum(kernel_values) / len(data)
    return np.mean(np.log(densities + 1e-9))

def true_local_kde(data, query_points, k, kernel_func):
    # Fit NearestNeighbors once
    nbrs = NearestNeighbors(n_neighbors=k).fit(data)
    distances, _ = nbrs.kneighbors(query_points)
    local_bandwidths = distances[:, -1]  # Bandwidth for each query point

    # Compute densities using the precomputed distances and bandwidths
    densities = np.zeros(len(query_points))
    for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
        if h > 0:  # Avoid division by zero
            diffs = data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(data)
    return densities

# Evaluate each parameter combination
results = []
for params in tqdm(ParameterGrid(param_grid), desc="Grid Search Progress", unit="params"):
    k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
    if kernel in ["silverman", "scott"]:
        log_likelihood = evaluate_kde(data, query_points, k=None, kernel=kernel, alpha=None)
        results.append(("global", kernel, "global", log_likelihood))
    else:
        log_likelihood = evaluate_kde(data, query_points, k, kernel, alpha)
        results.append((k, kernel, alpha, log_likelihood))

# Sort results by log-likelihood
results_sorted = sorted(results, key=lambda x: x[3], reverse=True)

# Convert results to a DataFrame for display
results_df = pd.DataFrame(results_sorted, columns=["k", "kernel", "alpha", "log_likelihood"])

# Display results
display(results_df)

# Print the best parameters
best_params = max(results, key=lambda x: x[3])
print("Best Parameters:", best_params)

################################################################################################
## plotting the best parameters kde local bandwidth 2D scatter plot

# Compute densities using the best parameters
best_k, best_kernel, best_alpha = best_params[:3]
kernel_funcs = {
    "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
    "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
    "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
    "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
    "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
}

if best_kernel in ["silverman", "scott"]:
    if best_kernel == "silverman":
        global_bandwidth = np.mean(silverman_bandwidth(data))
    elif best_kernel == "scott":
        global_bandwidth = np.mean(scott_bandwidth(data))
    densities = global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"])
else:
    kernel_func = kernel_funcs[best_kernel]
    densities = true_local_kde(data, query_points, k=best_k, kernel_func=lambda u: kernel_func(u) * best_alpha)

densities = densities.reshape(xx.shape)

# Plot the result
plt.figure(figsize=(12, 6))

# KDE Density Plot
plt.subplot(1, 2, 1)
plt.contourf(xx, yy, densities, levels=50, cmap='viridis')
plt.colorbar(label='Density')
plt.title(f"2D KDE with {best_kernel.capitalize()} Bandwidth (k={best_k}, alpha={best_alpha})")
plt.xlabel("X")
plt.ylabel("Y")

# Scatter Plot of Data Points
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], s=5, color='black')
plt.title("Scatter Plot of Generated Data Points")
plt.xlabel("X")
plt.ylabel("Y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import ParameterGrid
from sklearn.neighbors import NearestNeighbors
from tqdm import tqdm

class KDE:
    def __init__(self, data, user_percentages=None, kernel_types=None, alpha_values=None):
        self.data = data
        self.user_percentages = user_percentages if user_percentages is not None else [10]
        self.kernel_types = kernel_types if kernel_types is not None else ["gaussian"]
        self.alpha_values = alpha_values if alpha_values is not None else [2.0]
        self.xx, self.yy = None, None
        self.results_df = None
        self.best_params = None
        self.best_densities = None

    def _calculate_bandwidth(self, method):
        if method == "silverman":
            std = np.std(self.data, axis=0)
            iqr = np.subtract(*np.percentile(self.data, [75, 25], axis=0))
            return 0.9 * np.minimum(std, iqr / 1.34) * len(self.data) ** (-1 / (self.data.shape[1] + 4))
        elif method == "scott":
            std = np.std(self.data, axis=0)
            return std * len(self.data) ** (-1 / (self.data.shape[1] + 4))

    def _global_kde(self, query_points, bandwidth, kernel_func):
        densities = np.zeros(len(query_points))
        for i, q in enumerate(query_points):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / bandwidth
            kernel_values = kernel_func(scaled_diffs) / bandwidth
            densities[i] = np.sum(kernel_values) / len(self.data)
        return densities

    def _evaluate_kde(self, query_points, k, kernel, alpha=1.0):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        }

        kernel_func = kernel_funcs.get(kernel, None)
        if not kernel_func:
            raise ValueError(f"Unsupported kernel: {kernel}")

        nbrs = NearestNeighbors(n_neighbors=k).fit(self.data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * alpha

        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(self.data)

        return np.mean(np.log(densities + 1e-9))

    def fit(self):
        x_min, x_max = self.data[:, 0].min(), self.data[:, 0].max()
        y_min, y_max = self.data[:, 1].min(), self.data[:, 1].max()
        x_margin = 0.1 * (x_max - x_min)
        y_margin = 0.1 * (y_max - y_min)

        x_min -= x_margin
        x_max += x_margin
        y_min -= y_margin
        y_max += y_margin

        self.xx, self.yy = np.meshgrid(np.linspace(x_min, x_max, 50), np.linspace(y_min, y_max, 50))
        query_points = np.vstack([self.xx.ravel(), self.yy.ravel()]).T

        k_values = [int(len(self.data) * (p / 100)) for p in self.user_percentages]
        results = []

        for params in tqdm(ParameterGrid({"k": k_values, "kernel": self.kernel_types, "alpha": self.alpha_values}), desc="Grid Search Progress"):
            k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
            log_likelihood = self._evaluate_kde(query_points, k, kernel, alpha)
            results.append((k, kernel, alpha, log_likelihood))

        self.results_df = pd.DataFrame(results, columns=["k", "kernel", "alpha", "log_likelihood"]).sort_values(by="log_likelihood", ascending=False)
        self.best_params = self.results_df.iloc[0]

        # Compute densities for the best parameters and store them
        best_k = int(self.best_params["k"])
        best_kernel = self.best_params["kernel"]
        best_alpha = self.best_params["alpha"]

        kernel_func = kernel_funcs[best_kernel]
        nbrs = NearestNeighbors(n_neighbors=best_k).fit(self.data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * best_alpha

        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = self.data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(self.data)

        self.best_densities = densities.reshape(self.xx.shape)
        return self.results_df

    def plot(self):
        if self.best_params is None:
            raise RuntimeError("You need to call the `fit` method before plotting.")

        if self.best_densities is None:
            raise RuntimeError("KDE densities have not been computed. Check the `fit` method.")

        # Use the precomputed densities for plotting
        plt.figure(figsize=(12, 6))

        # KDE Density Plot
        plt.subplot(1, 2, 1)
        plt.contourf(self.xx, self.yy, self.best_densities, levels=50, cmap="viridis")
        plt.colorbar(label="Density")
        plt.title(f"2D KDE with {self.best_params['kernel'].capitalize()} Bandwidth")
        plt.xlabel("X")
        plt.ylabel("Y")

        # Scatter Plot of Data Points
        plt.subplot(1, 2, 2)
        plt.scatter(self.data[:, 0], self.data[:, 1], s=5, color="black")
        plt.title("Scatter Plot of Data Points")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.tight_layout()
        plt.show()


# test class
df = read_file('training_data.csv')
kde = KDE(df.values, user_percentages=[10], alpha_values=[2], kernel_types=['gaussian'])
kde.fit()
kde.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import pandas as pd
from sklearn.model_selection import ParameterGrid
from tqdm import tqdm

class KDEEvaluator:
    def __init__(self, file_path, user_percentages):
        self.data = self.read_file(file_path).values
        self.user_percentages = user_percentages
        self.k_values = [int(len(self.data) * (p / 100)) for p in user_percentages]
        self.param_grid = {
            "k": self.k_values,
            "kernel": ["gaussian"],
            "alpha": [2.0],
        }
        self.query_points, self.xx, self.yy = self.create_meshgrid()

    @staticmethod
    def read_file(file_path):
        return pd.read_csv(file_path)

    def create_meshgrid(self):
        x_min, x_max = self.data[:, 0].min(), self.data[:, 0].max()
        y_min, y_max = self.data[:, 1].min(), self.data[:, 1].max()
        x_margin = 0.1 * (x_max - x_min)
        y_margin = 0.1 * (y_max - y_min)

        x_min -= x_margin
        x_max += x_margin
        y_min -= y_margin
        y_max += y_margin

        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
        query_points = np.vstack([xx.ravel(), yy.ravel()]).T
        return query_points, xx, yy

    @staticmethod
    def silverman_bandwidth(data):
        std = np.std(data, axis=0)
        iqr = np.subtract(*np.percentile(data, [75, 25], axis=0))
        return 0.9 * np.minimum(std, iqr / 1.34) * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def scott_bandwidth(data):
        std = np.std(data, axis=0)
        return std * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def global_kde(data, query_points, bandwidth, kernel_func):
        densities = np.zeros(len(query_points))
        for i, q in enumerate(query_points):
            diffs = data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / bandwidth
            kernel_values = kernel_func(scaled_diffs) / bandwidth
            densities[i] = np.sum(kernel_values) / len(data)
        return densities

    @staticmethod
    def evaluate_kde(data, query_points, k, kernel, alpha=1.0):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }
        if kernel == "silverman":
            global_bandwidth = np.mean(KDEEvaluator.silverman_bandwidth(data))
            return np.mean(np.log(KDEEvaluator.global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
        if kernel == "scott":
            global_bandwidth = np.mean(KDEEvaluator.scott_bandwidth(data))
            return np.mean(np.log(KDEEvaluator.global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
        kernel_func = kernel_funcs.get(kernel, None)
        if not kernel_func:
            raise ValueError(f"Unsupported kernel: {kernel}")

        nbrs = NearestNeighbors(n_neighbors=k).fit(data)
        densities = np.zeros(len(query_points))
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * alpha
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(data)
        return np.mean(np.log(densities + 1e-9))

    @staticmethod
    def true_local_kde(data, query_points, k, kernel_func):
        nbrs = NearestNeighbors(n_neighbors=k).fit(data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1]
        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            if h > 0:
                diffs = data - q
                norms = np.linalg.norm(diffs, axis=1)
                scaled_diffs = norms / h
                kernel_values = kernel_func(scaled_diffs) / h
                densities[i] = np.sum(kernel_values) / len(data)
        return densities

    def run(self):
        results = []
        for params in tqdm(ParameterGrid(self.param_grid), desc="Grid Search Progress", unit="params"):
            k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
            if kernel in ["silverman", "scott"]:
                log_likelihood = self.evaluate_kde(self.data, self.query_points, k=None, kernel=kernel, alpha=None)
                results.append(("global", kernel, "global", log_likelihood))
            else:
                log_likelihood = self.evaluate_kde(self.data, self.query_points, k, kernel, alpha)
                results.append((k, kernel, alpha, log_likelihood))

        results_sorted = sorted(results, key=lambda x: x[3], reverse=True)
        results_df = pd.DataFrame(results_sorted, columns=["k", "kernel", "alpha", "log_likelihood"])
        best_params = max(results, key=lambda x: x[3])
        return results_df, best_params

    def plot_results(self, best_params):
        best_k, best_kernel, best_alpha = best_params[:3]
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }

        if best_kernel in ["silverman", "scott"]:
            if best_kernel == "silverman":
                global_bandwidth = np.mean(self.silverman_bandwidth(self.data))
            elif best_kernel == "scott":
                global_bandwidth = np.mean(self.scott_bandwidth(self.data))
            densities = self.global_kde(self.data, self.query_points, global_bandwidth, kernel_funcs["gaussian"])
        else:
            kernel_func = kernel_funcs[best_kernel]
            densities = self.true_local_kde(self.data, self.query_points, k=best_k, kernel_func=lambda u: kernel_func(u) * best_alpha)

        densities = densities.reshape(self.xx.shape)

        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.contourf(self.xx, self.yy, densities, levels=50, cmap='viridis')
        plt.colorbar(label='Density')
        plt.title(f"2D KDE with {best_kernel.capitalize()} Bandwidth (k={best_k}, alpha={best_alpha})")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.subplot(1, 2, 2)
        plt.scatter(self.data[:, 0], self.data[:, 1], s=5, color='black')
        plt.title("Scatter Plot of Generated Data Points")
        plt.xlabel("X")
        plt.ylabel("Y")

        plt.tight_layout()
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import pandas as pd
from sklearn.model_selection import ParameterGrid
from tqdm import tqdm
pd.set_option("display.float_format", "{:.30f}".format)

class KDEEvaluator:
    def __init__(self, file_path, user_percentages=None, kernel_types=None, alpha_values=None, evaluation_metric='log_likelihood'):
        self.data = self.read_file(file_path).values
        self.user_percentages = user_percentages if user_percentages else [5, 10, 15, 20]
        self.k_values = [int(len(self.data) * (p / 100)) for p in self.user_percentages]
        self.kernel_types = kernel_types if kernel_types else ["gaussian", "epanechnikov", "biweight", "triangular", "cosine", "silverman", "scott"]
        self.alpha_values = alpha_values if alpha_values else [0.5, 1.0, 2.0]
        self.evaluation_metric = evaluation_metric
        self.param_grid = {
            "k": self.k_values,
            "kernel": self.kernel_types,
            "alpha": self.alpha_values,
        }
        self.query_points, self.xx, self.yy = self.create_meshgrid()

    @staticmethod
    def read_file(file_path):
        return pd.read_csv(file_path)

    def create_meshgrid(self):
        x_min, x_max = self.data[:, 0].min(), self.data[:, 0].max()
        y_min, y_max = self.data[:, 1].min(), self.data[:, 1].max()
        x_margin = 0.1 * (x_max - x_min)
        y_margin = 0.1 * (y_max - y_min)

        x_min -= x_margin
        x_max += x_margin
        y_min -= y_margin
        y_max += y_margin

        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
        query_points = np.vstack([xx.ravel(), yy.ravel()]).T
        return query_points, xx, yy

    @staticmethod
    def silverman_bandwidth(data):
        std = np.std(data, axis=0)
        iqr = np.subtract(*np.percentile(data, [75, 25], axis=0))
        return 0.9 * np.minimum(std, iqr / 1.34) * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def scott_bandwidth(data):
        std = np.std(data, axis=0)
        return std * len(data) ** (-1 / (data.shape[1] + 4))

    @staticmethod
    def global_kde(data, query_points, bandwidth, kernel_func):
        densities = np.zeros(len(query_points))
        for i, q in enumerate(query_points):
            diffs = data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / bandwidth
            kernel_values = kernel_func(scaled_diffs) / bandwidth
            densities[i] = np.sum(kernel_values) / len(data)
        return densities

    @staticmethod
    def true_local_kde(data, query_points, k, kernel_func):
        nbrs = NearestNeighbors(n_neighbors=k).fit(data)
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1]
        densities = np.zeros(len(query_points))
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            if h > 0:  # Avoid division by zero
                diffs = data - q
                norms = np.linalg.norm(diffs, axis=1)
                scaled_diffs = norms / h
                kernel_values = kernel_func(scaled_diffs) / h
                densities[i] = np.sum(kernel_values) / len(data)
        return densities

    @staticmethod
    def evaluate_log_likelihood(data, query_points, k, kernel, alpha=1.0):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }
        if kernel == "silverman":
            global_bandwidth = np.mean(KDEEvaluator.silverman_bandwidth(data))
            return np.mean(np.log(KDEEvaluator.global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
        if kernel == "scott":
            global_bandwidth = np.mean(KDEEvaluator.scott_bandwidth(data))
            return np.mean(np.log(KDEEvaluator.global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))
        kernel_func = kernel_funcs.get(kernel, None)
        if not kernel_func:
            raise ValueError(f"Unsupported kernel: {kernel}")

        nbrs = NearestNeighbors(n_neighbors=k).fit(data)
        densities = np.zeros(len(query_points))
        distances, _ = nbrs.kneighbors(query_points)
        local_bandwidths = distances[:, -1] * alpha
        for i, (q, h) in enumerate(zip(query_points, local_bandwidths)):
            diffs = data - q
            norms = np.linalg.norm(diffs, axis=1)
            scaled_diffs = norms / h
            kernel_values = kernel_func(scaled_diffs) / h
            densities[i] = np.sum(kernel_values) / len(data)
        return np.mean(np.log(densities + 1e-9))

    @staticmethod
    def evaluate_mise(data, query_points, k, kernel, alpha=1.0):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        }
        estimated_densities = KDEEvaluator.global_kde(data, query_points, KDEEvaluator.silverman_bandwidth(data).mean(), kernel_funcs["gaussian"])
        return np.mean((1 - estimated_densities) ** 2)


    def evaluate(self, data, query_points, k, kernel, alpha):
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
        }

        if kernel in ["silverman", "scott"]:
            global_bandwidth = (
                np.mean(self.silverman_bandwidth(data)) if kernel == "silverman"
                else np.mean(self.scott_bandwidth(data))
            )
            return np.mean(np.log(self.global_kde(data, query_points, global_bandwidth, kernel_funcs["gaussian"]) + 1e-9))

        # For local bandwidths, use the evaluation metric
        if self.evaluation_metric == "log_likelihood":
            return self.evaluate_log_likelihood(data, query_points, k, kernel, alpha)
        elif self.evaluation_metric == "mise":
            return self.evaluate_mise(data, query_points, k, kernel, alpha)
        else:
            raise ValueError(f"Unsupported evaluation metric: {self.evaluation_metric}")


    def run(self):
        results = []
        evaluated_global_kernels = {}

        # Separate global kernels and local kernels
        global_kernels = [kernel for kernel in self.kernel_types if kernel in ["silverman", "scott"]]
        local_params = [
            params for params in ParameterGrid(self.param_grid) if params["kernel"] not in global_kernels
        ]

        # Evaluate global kernels
        for kernel in global_kernels:
            if kernel not in evaluated_global_kernels:
                score = self.evaluate(self.data, self.query_points, k=None, kernel=kernel, alpha=None)
                evaluated_global_kernels[kernel] = score
                results.append(("global", kernel, "global", score))  # Append only once

        # Evaluate local kernels with tqdm for progress
        for params in tqdm(local_params, desc="Grid Search Progress", unit="params"):
            k, kernel, alpha = params["k"], params["kernel"], params["alpha"]
            score = self.evaluate(self.data, self.query_points, k, kernel, alpha)
            results.append((k, kernel, alpha, score))

        results_sorted = sorted(results, key=lambda x: x[3], reverse=True)
        results_df = pd.DataFrame(results_sorted, columns=["k", "kernel", "alpha", "score"])
        if self.evaluation_metric not in ['mise']:
            best_params = max(results, key=lambda x: x[3])
        else:
            best_params = min(results, key=lambda x: x[3])
        return results_df, best_params



    def plot_results(self, best_params):
        best_k, best_kernel, best_alpha = best_params[:3]
        kernel_funcs = {
            "gaussian": lambda u: np.exp(-0.5 * u**2) / np.sqrt(2 * np.pi),
            "epanechnikov": lambda u: 0.75 * (1 - u**2) * (u <= 1),
            "biweight": lambda u: (15 / 16) * (1 - u**2)**2 * (u <= 1),
            "triangular": lambda u: (1 - np.abs(u)) * (np.abs(u) <= 1),
            "cosine": lambda u: (np.pi / 4) * np.cos(np.pi * u / 2) * (np.abs(u) <= 1),
        }

        if best_kernel in ["silverman", "scott"]:
            if best_kernel == "silverman":
                global_bandwidth = np.mean(self.silverman_bandwidth(self.data))
            elif best_kernel == "scott":
                global_bandwidth = np.mean(self.scott_bandwidth(self.data))

            densities = self.global_kde(self.data, self.query_points, global_bandwidth, kernel_funcs["gaussian"])
            scatter_densities = self.global_kde(self.data, self.data, global_bandwidth, kernel_funcs["gaussian"])
        else:
            kernel_func = kernel_funcs[best_kernel]
            densities = self.true_local_kde(self.data, self.query_points, k=best_k, kernel_func=lambda u: kernel_func(u) * best_alpha)
            scatter_densities = self.true_local_kde(self.data, self.data, k=best_k, kernel_func=lambda u: kernel_func(u) * best_alpha)

        densities = densities.reshape(self.xx.shape)

        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.contourf(self.xx, self.yy, densities, levels=50, cmap='viridis')
        plt.colorbar(label='Density')
        plt.title(f"2D KDE with {best_kernel.capitalize()} Bandwidth (k={best_k}, alpha={best_alpha})")
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.xlim(self.xx.min(), self.xx.max())
        plt.ylim(self.yy.min(), self.yy.max())

        plt.subplot(1, 2, 2)
        plt.scatter(self.data[:, 0], self.data[:, 1], c=scatter_densities, cmap='viridis', s=5)
        plt.colorbar(label='Density')
        plt.title("Scatter Plot of Data Points with KDE Densities")
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.xlim(self.xx.min(), self.xx.max())
        plt.ylim(self.yy.min(), self.yy.max())

        plt.tight_layout()
        plt.show()
```

```python
# Define the path to your dataset
file_path = "training_data.csv"

# Define user percentages for k values

# Initialize the KDEEvaluator instance
kde_evaluator = KDEEvaluator(file_path=file_path, user_percentages=None,kernel_types=['scott'],
    alpha_values=None,evaluation_metric=None)

# Run the KDE evaluation
results_df, best_params = kde_evaluator.run()

# Display the results
print("Evaluation Results:")
print(results_df)

print("\nBest Parameters:")
print(best_params)

# Plot the results with the best parameters
kde_evaluator.plot_results(best_params)
```
